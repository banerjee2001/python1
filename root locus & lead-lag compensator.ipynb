```python
import os
import matplotlib.pyplot as plt 
from control.matlab import *
```

```python
# Root locus without compensator.

s = TransferFunction.s
k=1
G = k/(s*(s+1)*(s+2))
sys = tf(G)
locus = rlocus(sys)
plt.show(locus)
```

```python
# Lead compensator.

s = TransferFunction.s
kc=0.966
G = 1/(s*(s+1)*(s+2))
Gc = kc*(2.33+s)
sys = tf(G*Gc)
locus1 = rlocus(sys)
plt.show(locus1)
```

```python
# Lag compensator.

s = TransferFunction.s
kc=0.966
G = 1/(s*(s+1)*(s+2))
Gc = kc/(s+0.5)
sys = tf(G*Gc)
locus1 = rlocus(sys)
plt.show(locus1)
```
